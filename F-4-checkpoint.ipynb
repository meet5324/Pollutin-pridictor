## Introduction
This project involves analyzing student performance data to explore how demographic and educational factors influence academic success. The dataset includes student scores in math, reading, and writing, along with categorical information such as gender, race/ethnicity, parental education level, lunch type, and test preparation status. The aim is to perform in-depth data exploration, generate visual insights, and classify students based on their performance levels using a custom target variable.

## Problem Statement
Educational institutions often struggle to identify students at risk of underperforming. Without targeted interventions, these students may not reach their full potential. The objective of this project is to:

Analyze the relationship between demographic attributes and student performance.

Create a classification model to categorize students as high or low performers.

Visualize feature interactions to uncover patterns and influential factors.

```python
#Step 1: Load Libraries and Dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("StudentsPerformance.csv")  # Replace with your actual file path
df
```

The dataset `StudentsPerformance.csv` is successfully loaded into a pandas DataFrame named `df` for further analysis.

```python
#Step 2: Basic Data Exploration
print(df.head())
print(df.info())
print(df.describe(include='all'))
print(df.isnull().sum())
```

The dataset has been explored to show its first few rows, data types, summary statistics for all columns, and verified to have no missing values.

```python
#Step 3: Create Target Variable for Classification
df['average_score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)

# Create binary target
df['high_performer'] = (df['average_score'] >= 70).astype(int)

# Class distribution
print(df['high_performer'].value_counts())
sns.countplot(x='high_performer', data=df)
plt.title('Target Variable Distribution (High Performer)')
plt.show()
```

A new target variable `high_performer` was created based on average scores, classifying students scoring 70 or above as high performers, with the class distribution visualized to show the balance between categories.

```python
#Step 4: Correlation Analysis
num_cols = ['math score', 'reading score', 'writing score', 'average_score']

# Correlation matrix
corr = df[num_cols].corr()

# Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

The correlation analysis reveals strong positive relationships among math, reading, writing scores, and the average score, confirming these metrics are closely related.

```python
#Step 5: Interaction Between Categorical Features and Target
categorical_cols = ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']

for col in categorical_cols:
    plt.figure(figsize=(8,4))
    sns.countplot(x=col, hue='high_performer', data=df)
    plt.title(f'High Performer Distribution by {col}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

The analysis shows how the distribution of high performers varies across different categories like gender, race/ethnicity, parental education, lunch type, and test preparation, highlighting potential demographic and socio-economic influences on performance.

```python
# Step 6: Boxplot for Feature vs Average Score
for col in categorical_cols:
    plt.figure(figsize=(8,4))
    sns.boxplot(x=col, y='average_score', hue='high_performer', data=df)
    plt.title(f'Average Score by {col} and Performance')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

Boxplots reveal how average scores differ within categorical groups and between high and low performers, highlighting distinct performance patterns linked to demographic and educational factors.

```python
#Step 7: Pairplot (Optional but Insightful)
sns.pairplot(df, hue='high_performer', vars=['math score', 'reading score', 'writing score'])
plt.suptitle("Pairplot of Scores by Performance", y=1.02)
plt.show()
```

The pairplot visualizes score distributions and relationships across subjects, clearly distinguishing high and low performers and revealing patterns in studentsâ€™ academic strengths.
